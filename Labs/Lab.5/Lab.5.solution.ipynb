3. Write functions that create special matrices (note these are standalone functions, not member functions of your `matrix` class):
    * `constant(n,m,c)`: returns a `n` by `m` matrix filled with floats of value `c`.
    * `zeros(n,m)` and `ones(n,m)`: return `n` by `m` matrices filled with floats of value `0` and `1`, respectively.
    * `eye(n)`: returns the n by n identity matrix.

```python
def constant(n, m, c):
    return matrix([[float(c) for _ in range(m)] for _ in range(n)])
def zeros(n, m):
    return constant(n, m, 0)
def ones(n, m):
    return constant(n, m, 1)
def eye(n):
    identity = [[1.0 if x == y else 0.0 for y in range(n)] for x in range(n)]
    return matrix(identity)
```

4. Add the following member functions to your class. Make sure to appropriately test the dimensions of the matrices to make sure the operations are correct.
    * `M.scalarmul(c)`: a matrix that is scalar product $cM$, where every element of $M$ is multiplied by $c$.
    * `M.add(N)`: adds two matrices $M$ and $N$. Donâ€™t forget to test that the sizes of the matrices are compatible for this and all other operations.
    * `M.sub(N)`: subtracts two matrices $M$ and $N$.
    * `M.mat_mult(N)`: returns a matrix that is the matrix product of two matrices $M$ and $N$.
    * `M.element_mult(N)`: returns a matrix that is the element-wise product of two matrices $M$ and $N$.
    * `M.equals(N)`: returns true/false if $M==N$.

```python
class matrix:
    def __init__(self, *args):
        if len(args) == 2 and isinstance(args[0], int) and isinstance(args[1], int):
            self.n = args[0]
            self.m = args[1]
            self.ar = [[0 for _ in range(self.m)] for _ in range(self.n)]
        elif len(args) == 1 and isinstance(args[0], list):
            self.ar = args[0]
            self.n = len(self.ar)
            if self.n > 0:
                self.m = len(self.ar[0])
                if not all(len(row) == self.m for row in self.ar):
                    print("Rows don't match the columns.")
            else:
                self.m = 0
        else:
            print("Invalid initialization.")

    def __getitem__(self, ind):
        if isinstance(ind, tuple):
            x, y = ind
            return self.ar[x][y]
        else:
            return self.ar[ind]
    def __setitem__(self, ind, val):
        if isinstance(ind, tuple):
            x, y = ind
            self.ar[x][y] = val
        else:
            self.ar[ind] = val
    def __str__(self):
        return "\n".join([" ".join(map(str, row)) for row in self.ar])
    def shape(self):
        return (self.n, self.m)
    def transpose(self):
        transposed = [[self.ar[y][x] for y in range(self.n)] for x in range(self.m)]
    def row(self, n):
        if n >= self.n and n < 0:
            print("Out of range")
        else:
            return matrix([self.ar[n]])
    def column(self, n):
        if n >= self.m and n < 0:
            print("Out of range")
        else:
            return matrix([[self.ar[x][n]] for x in range(self.n)])
    def to_list(self):
        return [row[:] for row in self.ar]
    def block(self, n_0, n_1, m_0, m_1):
        blockArray = [self.ar[x][m_0:m_1+1] for x in range(n_0, n_1+1)]
        return matrix(blockArray)
    def scalarmul(self, c):
        scaled = [[self.ar[x][y] * c for y in range(self.m)] for x in range(self.n)]
        return matrix(scaled)
    def add(self, N):
        if N.shape() != self.shape():
            print("Must have the same shape to add")
        else:
            result = [[self.ar[x][y] + N[x][y] for y in range(self.m)] for x in range(self.n)]
            return matrix(result)
    def sub(self, N):
        if N.shape() != self.shape():
            print("Must have the same shape to add")
        else:
            result = [[self.ar[x][y] - N[x][y] for y in range(self.m)] for x in range(self.n)]
            return matrix(result)
    def mat_mult(self, N):
        if N.n != self.m:
            print("Number of columns and rows do not match.")
        else:
            result = [[sum(self.ar[x][z] * N[z][y] for z in range(self.m)) for y in range(N.m)] for x in range(self.n)]
            return matrix(result)
    def element_mult(self, N):
        if N.shape() != self.shape():
            print("Not the same shape.")
        else:
            result = [[self.ar[x][y] * N[x][y] for y in range(self.m)] for x in range(self.n)]
            return matrix(result)
    def equals(self, N):
        if N.shape() != self.shape():
            return False
        else:
            return all(self.ar[x][y] == N[x][y] for x in range(self.n) for y in range(self.m))

m1 = matrix([[1,2], [3,4]])
m2 = matrix([[6,7], [8,9]])

print("Scalar multiplication:\n",m1.scalarmul(2))
print("Matrix addition:\n", m1.add(m2))
print("Matrix subtraction:\n", m1.sub(m2))
print("Matrix multiplication:\n", m1.mat_mult(m2))
print("Matrix element multiplication:\n", m1.element_mult(m2))
print("Is m1 equal to m1: ", m1.equals(m1))
print("Is m1 equal to m2: ", m1.equals(m2))
```
