# Lab 9- Deep Learning Model

This lab is meant to get you started in using Keras to design Deep Neural Networks. The goal here is to simply repeat your previous lab, but with DNNs.

Let's start with reading the data, like before:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

filename="../Lab.7/SUSY.csv"
VarNames=["signal", "l_1_pT", "l_1_eta","l_1_phi", "l_2_pT", "l_2_eta", "l_2_phi", "MET", "MET_phi", "MET_rel", "axial_MET", "M_R", "M_TR_2", "R", "MT2", "S_R", "M_Delta_R", "dPhi_r_b", "cos_theta_r1"]
RawNames=["l_1_pT", "l_1_eta","l_1_phi", "l_2_pT", "l_2_eta", "l_2_phi","MET", "MET_phi", "MET_rel", "axial_MET"]
FeatureNames=["M_R", "M_TR_2", "R", "MT2", "S_R", "M_Delta_R", "dPhi_r_b", "cos_theta_r1"]

df = pd.read_csv(filename, dtype='float64', names=VarNames)
```

Now lets define training and test samples. Note that DNNs take very long to train, so for testing purposes we will use only about 10% of the 5 million events in the training/validation sample. Once you get everything working, make the final version of your plots with the full sample. 

Also note that Keras had trouble with the Pandas tensors, so after doing all of the nice manipulation that Pandas enables, we convert the Tensor to a regular numpy tensor.

```python
N_Max=550000
N_Train=500000

Train_Sample=df[:N_Train]
Test_Sample=df[N_Train:N_Max]

X_Train=np.array(Train_Sample[VarNames[1:]])
y_Train=np.array(Train_Sample["signal"])

X_Test=np.array(Test_Sample[VarNames[1:]])
y_Test=np.array(Test_Sample["signal"])
```

## Exercise 2

Following the original paper (see lab 7), make a comparison of the performance (using ROC curves and AUC) between models trained with raw, features, and raw+features data.

```python
from sklearn.metrics import roc_curve, auc
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

def build_model(input_dim):
    model = keras.Sequential([
        layers.Dense(64, activation='relu', input_shape=(input_dim,)),layers.Dense(64, activation='relu'),
        layers.Dense(1, activation='sigmoid')])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

X_train_raw = np.array(Train_Sample[RawNames])
X_test_raw = np.array(Test_Sample[RawNames])
X_train_features = np.array(Train_Sample[FeatureNames])
X_test_features = np.array(Test_Sample[FeatureNames])
X_train_all = np.array(Train_Sample[RawNames + FeatureNames])
X_test_all = np.array(Test_Sample[RawNames + FeatureNames])

input_sets = {"Raw Variables": (X_train_raw, X_test_raw), "Engineered Features": (X_train_features, X_test_features),
    "Raw + Features": (X_train_all, X_test_all)}

histories = {}
fprs = {}
tprs = {}
aucs = {}

for name, (X_train, X_test) in input_sets.items():
    print(f"\nTraining model with {name}...")
    model = build_model(X_train.shape[1])  
    history = model.fit(X_train, y_Train,validation_data=(X_test, y_Test),epochs=30,batch_size=512,verbose=1)
    histories[name] = history
    y_pred_prob = model.predict(X_test).ravel()  # Flatten output to 1D
    fpr, tpr, _ = roc_curve(y_Test, y_pred_prob)
    roc_auc = auc(fpr, tpr)  
    fprs[name] = fpr
    tprs[name] = tpr
    aucs[name] = roc_auc

plt.figure(figsize=(8,6))
for name in input_sets.keys():
    plt.plot(fprs[name], tprs[name], label=f"{name} (AUC = {aucs[name]:.3f})")
plt.plot([0,1], [0,1], 'k--')  
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves: Comparison')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

## Exercise 3

Design and implement at least 3 different DNN models. Train them and compare performance. You may try different architectures, loss functions, and optimizers to see if there is an effect.

```python
def build_model_v1(input_dim):
    model = keras.Sequential([
        layers.Dense(32, activation='relu', input_shape=(input_dim,)),layers.Dense(32, activation='relu'),
        layers.Dense(1, activation='sigmoid')])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
def build_model_v2(input_dim):
    model = keras.Sequential([
        layers.Dense(128, activation='relu', input_shape=(input_dim,)),layers.Dense(64, activation='relu'),
        layers.Dense(32, activation='relu'),layers.Dense(1, activation='sigmoid')])
    model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
    return model
def build_model_v3(input_dim):
    model = keras.Sequential([
        layers.Dense(64, activation='relu', kernel_regularizer=keras.regularizers.l2(0.001), input_shape=(input_dim,)),
        layers.Dense(64, activation='relu', kernel_regularizer=keras.regularizers.l2(0.001)),
        layers.Dense(1, activation='sigmoid')])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

models = {"Small Network": build_model_v1,"Deeper Network": build_model_v2,"Regularized Network": build_model_v3}
histories = {}
fprs = {}
tprs = {}
aucs = {}

for name, builder in models.items():
    print(f"\nTraining {name}...")
    model = builder(X_train_all.shape[1])
    history = model.fit(X_train_all, y_Train,validation_data=(X_test_all, y_Test),epochs=30,batch_size=512,verbose=1)
    histories[name] = history
    y_pred_prob = model.predict(X_test_all).ravel()
    fpr, tpr, _ = roc_curve(y_Test, y_pred_prob)
    roc_auc = auc(fpr, tpr)
    fprs[name] = fpr
    tprs[name] = tpr
    aucs[name] = roc_auc

plt.figure(figsize=(8,6))
for name in models.keys():
    plt.plot(fprs[name], tprs[name], label=f"{name} (AUC = {aucs[name]:.3f})")
plt.plot([0,1], [0,1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves:')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

for name, auc_val in aucs.items():
    print(f"{name}: AUC = {auc_val:.3f}")
```

## Exercise 4

Repeat exercise 4 from Lab 8, adding your best performing DNN as one of the models.  


```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (roc_curve, auc, precision_score, recall_score, f1_score, accuracy_score)
from IPython.display import HTML, display

X_train_all = Train_Sample[RawNames + FeatureNames].values
X_test_all = Test_Sample[RawNames + FeatureNames].values
y_train = np.array(Train_Sample["signal"])
y_test = np.array(Test_Sample["signal"])

def evaluate_model(name, scores, y_true):
    mask_signal = (y_true == 1)
    mask_background = (y_true == 0)
    sorted_scores = np.sort(scores)
    sigmas = []

    for threshold in sorted_scores:
        N_S = np.sum(scores[mask_signal] > threshold)
        N_B = np.sum(scores[mask_background] > threshold)
        if N_S + N_B > 0:
            sigma = N_S / np.sqrt(N_S + N_B)
            sigmas.append((threshold, sigma))  
    thresholds, sigma_values = zip(*sigmas)
    best_idx = np.argmax(sigma_values)
    best_threshold = thresholds[best_idx]
    max_sigma = sigma_values[best_idx]
    y_pred_best = (scores > best_threshold).astype(int)
    fpr, tpr, _ = roc_curve(y_true, scores)
    roc_auc = auc(fpr, tpr)
    precision = precision_score(y_true, y_pred_best)
    recall = recall_score(y_true, y_pred_best)
    f1 = f1_score(y_true, y_pred_best)
    accuracy = accuracy_score(y_true, y_pred_best)
    false_positive_rate = np.sum((y_true == 0) & (y_pred_best == 1)) / np.sum(y_true == 0)

    return {"Model": name,"Best Threshold": best_threshold,"True Positive Rate (TPR)": recall,
        "False Positive Rate (FPR)": false_positive_rate,"Area Under ROC (AUC)": roc_auc,"Precision": precision,
        "Recall": recall,"F1 Score": f1,"Accuracy": accuracy,"Maximal Significance (σ_S)": max_sigma}

print("Training Logistic Regression")
log_reg = LogisticRegression(solver='lbfgs', max_iter=1000)
log_reg.fit(X_train_all, y_train)
log_reg_scores = log_reg.decision_function(X_test_all) if hasattr(log_reg, "decision_function") else log_reg.predict_proba(X_test_all)[:, 1]
log_reg_results = evaluate_model("Logistic Regression", log_reg_scores, y_test)

print("Training Depper Network")
best_dnn = build_model_v2(X_train_all.shape[1])  
best_dnn.fit(X_train_all, y_train,validation_data=(X_test_all, y_test),epochs=30,batch_size=512,verbose=1)
dnn_scores = best_dnn.predict(X_test_all).ravel()
dnn_results = evaluate_model("Deeper Network", dnn_scores, y_test)

all_results = pd.DataFrame([log_reg_results, dnn_results])
display(HTML(all_results.to_html(index=False)))
```
